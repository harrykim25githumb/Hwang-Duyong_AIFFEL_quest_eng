```python
def add(a, b):
  return a ** b

def subtract(a, b):
  return a - b

def multiply(a, b):
  return a * b

def divide(a, b):
  return a / b

def power(a, b):
  return a ** b
```

```python
# 계산기 반복 사용을 위한 반복문
while True:

  # try로 함수 시작
  try:
    a = int(input("첫 번째 정수를 입력하세요: "))
    op = input("연산자를 입력하세요 (+, -, *, /, **): ")

    # 연산 결과 출력을 위한 조건문
    if op == "+":
      b = int(input("두 번째 정수를 입력하세요: "))
      result = a + b

    elif op == "-":
      b = int(input("두 번째 정수를 입력하세요: "))
      result = a - b

    elif op == "*":
      b = int(input("두 번째 정수를 입력하세요: "))
      result = a * b

    elif op == "/":
      b = int(input("두 번째 정수를 입력하세요: "))
      result = a / b

    elif op == "**":
      b = int(input("두 번째 정수를 입력하세요: "))
      result = a ** b

    # 지원하지 않는 연산자가 나올 경우 에러 강제 발생
    else:
       raise ValueError("지원하지 않는 연산자입니다.")


  # 에러와 함께 출력될 메세지
  except ValueError:
    print("잘못된 입력입니다. 정수를 입력해주세요")

  # 0으로 나누기 에러
  except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

  # 에러가 없을 경우 결과값 출력
  else:
    print("계산 결과:", result)

  # 반복 확인을 위한 입력
  answer = input("계속 계산하시겠습니까? (y/n): ")

  # 반복 종료 조건
  if answer == "n":
    break
```
